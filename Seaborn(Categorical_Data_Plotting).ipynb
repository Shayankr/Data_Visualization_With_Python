```python
# x-axis --> Categorical Data
# y-axis --> Numerical Data
# so, it is a plotting of (or, variation in) numerical data with respect to the categorical variable

sns.catplot(x='day', y='total_bill', data=tips)
```

```python
# change theme setting plot the above graph again

sns.set(style='darkgrid')
sns.catplot(x='day', y='total_bill', data=tips)
```

Now, again change the theme of graph

```python
sns.set_theme(style='ticks')
```

```python
sns.catplot(x='day',y='total_bill', data=tips)
```

```python
# Drawn horizontal graph by interchanging x and y data.
sns.catplot(x='total_bill', y='day', data=tips)
```

Jitter is simply the addition of a small amount of horizontal (or vertical) variability to the data in order to ensure all data points are visible. 

```python
# jiiter=False
sns.catplot(x='day', y='total_bill', data=tips, jitter=False)  #kind='strip' by default.
```

```python
# kind='swarm'  --> a kind of Tree
sns.catplot(x='day', y='total_bill', data=tips, kind='swarm')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='swarm',
            hue='sex')
```

```python
sns.catplot(x='smoker', y='tip', data=tips, )
```

```python
# Changing the order of the categorical variable on x-axis.
sns.catplot(x='smoker', y='tip',data=tips, order=['No','Yes'])
```

```python
# Partiotion graph column-wise based on 'smoker'
sns.catplot(x='day', y='total_bill',data=tips, col='smoker')
```

```python
# Change the order of column
sns.catplot(x='day', y='total_bill',data=tips,hue='sex', col='smoker', col_order=['No',"Yes"]) # also add hue='sex'
```

Use a different height and aspect ratio for the facets:

```python
sns.catplot(x='day', y='total_bill',data=tips, col='smoker', col_order=['No',"Yes"],
            height=6, aspect=1.2, hue='sex')
```

```python
sns.swarmplot(x="day", y="total_bill", data=tips)
```

```python
sns.swarmplot(x="total_bill", y="day", data=tips)
```

Color the points using a second categorical variable:

```python
sns.swarmplot(x="day", y="total_bill", hue="sex", data=tips)
```

Split each level of the hue variable along the categorical axis:

```python
sns.swarmplot(x="day", y="total_bill", hue="smoker", 
              data=tips, palette="Set2", dodge=True)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.swarmplot(x="day", y="total_bill", hue="smoker", 
              data=tips, dodge=True, )
plt.show()
```

```python
sns.swarmplot(x="time", y="total_bill", data=tips, order=["Dinner", "Lunch"])
```

# Categorical Estimate Plots

**pointplot()**

* (with kind="point")*

pointplot() : This method is used to show point estimates and confidence intervals using scatter plot glyphs. A point plot represents an estimate of central tendency for a numeric variable by the position of scatter plot points and provides some indication of the uncertainty around that estimate using error bars.

```python
# We can change the theme
sns.set_theme(style="darkgrid")
```

```python
# pointplot
sns.pointplot(x="time", y="total_bill", data=tips)
```

```python
# Draw a set of vertical points with nested grouping by a two variables:
sns.pointplot(x="time", y="total_bill", hue="smoker",
                   data=tips)
```

```python
# Separate the points for different hue levels along the categorical axis:

sns.pointplot(x="time", y="total_bill", hue="smoker", data=tips, dodge=True)
```

Use a different marker and line style for the hue levels:

```python
sns.pointplot(x="time", y="total_bill", hue="smoker", data=tips,
              markers=["o", "x"],
              linestyles=["-", "--"])
```

Draw a set of horizontal points:

```python
sns.pointplot(x="tip", y="day", data=tips)
```

Donât draw a line connecting each point:

```python
sns.pointplot(x="tip", y="day", data=tips, join=False)
```

Use a different color palette for the points:

```python
sns.pointplot(x="time", y="total_bill", hue="smoker", data=tips, 
              palette="Set2")
```

Use median as the estimate of central tendency:

```python
# from numpy import median
```

```python
sns.pointplot(x="day", y="tip", data=tips, estimator=np.median)
```

Show the standard error of the mean with the error bars:

```python
sns.pointplot(x="day", y="tip", data=tips, ci=60)
```

Now, representing standard deviation of observations instead of a confidence interval

```python
sns.pointplot(x="day", y="tip", data=tips, ci="sd")
```

Now, Adding âcapsâ to the error bars

```python
sns.pointplot(x="day", y="tip", data=tips, capsize=.2)
```

Now,

```python
sns.catplot(x="sex", y="total_bill", hue="smoker", 
            col="time", data=tips, 
            kind="point",
            dodge=True,
            height=6, aspect=.8);
```
