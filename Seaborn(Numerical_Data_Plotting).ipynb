Visualization can be a core-component of this process
because when data are visualized properly, the human visual system can see trends and patterns that indicate a relationship.

Now, load 'fMRI' datset

Functional magnetic resonance imaging or functional MRI (fMRI) measures brain activity by detecting changes associated with blood flow. This technique relies on the fact that cerebral blood flow and neuronal activation are coupled.

<img src="https://upload.wikimedia.org/wikipedia/commons/4/4a/1206_FMRI.jpg" height=150, width=150/>

```python
fmri = sns.load_dataset('fmri')
```

```python
# look at first 5 rows:
fmri.head()
```

```python
# dimensions of dataset
fmri.shape
```

```python
sns.relplot(x='timepoint', y='signal', kind='line', data=fmri, )

# It also shows a line computed with 95% confident. (confidence-interval)--> S.D.
```

```python
np.std(fmri)
```

```python
sns.relplot(x='timepoint', y='signal', kind='line', data=fmri, ci=False)
```

```python
# change the err_style to 'bars'
sns.lineplot(x='timepoint', y='signal', hue='region', data=fmri,
             style='event', markers=True, ci=68, err_style='bars')
```

```python
# query 'fmri' dataset based on region=='frontal'
sns.lineplot(x='timepoint', y='signal', hue='event', units='subject', estimator=None, 
             lw=1, data= fmri.query(" region == 'frontal' "))
```
